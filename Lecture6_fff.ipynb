Exercise 2: Read the climate.csv file to a variable named 'climate', generate a line plot from 'Temperature (Celsius)', in the next step change the index to Country and in the 3rd step create a plot with all 3 variables and set the index to the Date. (put the title and axes labels)

```python
climate = pd.read_csv('climate.csv')
climate
```

```python
climate.set_index("Country")["Temperature (Celsius)"].plot()
```

```python
climate = climate.set_index("Date")
climate
```

```python
climate[:].plot();
plt.xticks(rotation=45, ha="right");
```

```python
climate[["Temperature (Celsius)", "Humidity (%)", "Wind Speed (km/h)"]].plot();
plt.xticks(rotation=45, ha="right");
```

```python
climate.set_index("Date")[["Temperature (Celsius)", "Humidity (%)", "Wind Speed (km/h)"]].plot(color=("green", "blue", "red"), linewidth=0.5);
plt.xticks(rotation=45, ha="right");
plt.ylabel("values", fontsize=16);
plt.xlabel("date", fontsize=16);
plt.tick_params(axis="both", labelsize=8);
plt.legend();
#label der x achse werden um 45 grad rotiert, horizontal alignment (ha) um den Namen genau an die Position zu bringen
#plt.tick_params(axis="both", labelsize=8) ändert die größe der Achsen-Beschriftungen
```

```python
climate = pd.read_csv('climate.csv')
climate = climate.set_index("Date")
climate["Temperature (Celsius)"].plot()
date = climate.index.tolist()
positions = range(len(date))
plt.xticks(positions, date, rotation=70, ha="right")
plt.show()

#um alle Werte auf die x-Achse zu bekommen
```
