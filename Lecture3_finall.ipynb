Exercise4: Write a Python program to find the median of a given array of numbers with for and if loop

```python
b = np.array=[2, 4, 5, 6, 8, 9]
array_len = len(b)

if array_len == 0:
    median = (b[int(array_len/2)] + b[int(array_len/2-1)]/2)
else: 
    median = b[int(array_len/2)]
print(median)

    
```
